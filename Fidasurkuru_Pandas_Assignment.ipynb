```python
#Cleaning THE Data
#Return a new Data Frame with no empty cells
df = pd.read_csv('data.csv')

new_df = df.dropna()

print(new_df.to_string())
```

```python
#Remove all rows with NULL values
df.dropna(inplace=True)
df
```

```python
#Replace NULL values with the number 130

df.fillna(130, inplace = True)
df
```

```python
#Replace NULL values in the "Calories" columns with the number 130
df["Calories"].fillna(130,inplace=True)
```

```python
df
```

```python
#Calculate the mean, and replace any empty values 
x=df["Calories"].mean()
df["Calories"].fillna(x,inplace=True)
df
```

```python
#Calculate the median, and replace any empty values
df=pd.read_csv('data.csv')
x=df["Calories"].median() 
x
```

```python
#Calculate the mode, and replace any empty values
x=df["Calories"].mode()
x
```

```python
#Convert to date
df['Date']=pd.to_datetime(df['Date'])
df
```

```python
#Remove rows with a NULL value in the "Date" column
df.dropna(subset=['Date'],inplace=True)
df
```

```python
#Set "Duration" = 45 in row 7
df.loc[7,'Duration']=45
df
```

```python
for x in df.index:
    if df.loc[x,"Duration"]>60:
        df.loc[x,"Duration"]=60
```

```python
df
```

```python
#Delete rows where "Duration" is higher than 60
for x in df.index:
    if df.loc[x,"Duration"]>60:
        df.drop(x,inplace=True)
        
```

```python
df
```

```python
#Returns True for every row that is a duplicate, othwerwise False
print(df.duplicated())
```

```python
#Remove all duplicates
df.drop_duplicates(inplace=True)
df
```
